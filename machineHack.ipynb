```python
import numpy as np
import pandas as pd
```

```python
dfTrain = pd.read_excel('Data_train.xlsx')
dfTest = pd.read_excel('Data_test.xlsx')
dfTrain
```

```python
dfTest
```

```python
dfTrain['Rating'].replace(to_replace =["NEW","Opening Soon","Temporarily Closed"],
                            value =0,inplace = True)
dfTrain['Rating'].replace(to_replace =["-"],
                            value =-999999,inplace = True)
dfTrain['Rating'] = pd.to_numeric(dfTrain['Rating'])
median = dfTrain.loc[dfTrain['Rating'] > 0, 'Rating'].median()
dfTrain["Rating"] = np.where(dfTrain["Rating"] <= 0, median,dfTrain['Rating'])
```

```python
dfTest['Rating'].replace(to_replace =["NEW","Opening Soon","Temporarily Closed"],
                            value =0,inplace = True)
dfTest['Rating'].replace(to_replace =["-"],
                            value =-999999,inplace = True)
dfTest['Rating'] = pd.to_numeric(dfTest['Rating'])
median = dfTest.loc[dfTest['Rating'] > 0, 'Rating'].median()
dfTest["Rating"] = np.where(dfTest["Rating"] <= 0, median,dfTest['Rating'])
```

```python
dfTrain['Votes'].replace(to_replace =["-"],
                            value =-999999,inplace = True)
dfTrain['Votes'] = pd.to_numeric(dfTrain['Votes'])

median = dfTrain.loc[dfTrain['Votes'] > 0, 'Votes'].median()
dfTrain["Votes"] = np.where(dfTrain["Votes"] <= 0, median,dfTrain['Votes'])
```

```python
dfTest['Votes'].replace(to_replace =["-"],
                            value =-999999,inplace = True)
dfTest['Votes'] = pd.to_numeric(dfTest['Votes'])

median = dfTest.loc[dfTest['Votes'] > 0, 'Votes'].median()
dfTest["Votes"] = np.where(dfTest["Votes"] <= 0, median,dfTest['Votes'])
```

```python
dfTrain['Reviews'].replace(to_replace =["-"],
                            value =-999999,inplace = True)
dfTrain['Reviews'] = pd.to_numeric(dfTrain['Reviews'])

median = dfTrain.loc[dfTrain['Reviews'] > 0, 'Votes'].median()
dfTrain["Reviews"] = np.where(dfTrain["Reviews"] <= 0, median,dfTrain['Reviews'])
```

```python
dfTest['Reviews'].replace(to_replace =["-"],
                            value =-999999,inplace = True)
dfTest['Reviews'] = pd.to_numeric(dfTest['Reviews'])

median = dfTest.loc[dfTest['Reviews'] > 0, 'Votes'].median()
dfTest["Reviews"] = np.where(dfTest["Reviews"] <= 0, median,dfTest['Reviews'])
```

```python
dfTrain['Minimum_Order'] = dfTrain['Minimum_Order'].apply(lambda x:int(x.strip('₹')))
dfTrain['Average_Cost'] = dfTrain['Average_Cost'].apply(lambda x:(x.strip('₹')))
dfTrain['Average_Cost'] = dfTrain['Average_Cost'].apply(lambda x:int(x.replace(',','')) if x != 'for' else -99)
median = int(dfTrain['Average_Cost'].median())
dfTrain["Average_Cost"] = np.where(dfTrain["Average_Cost"] == 0, median,dfTrain['Average_Cost'])

dfTrain['Delivery_Time'] = dfTrain['Delivery_Time'].apply(lambda x:int(x.strip(' minutes')))
```

```python
dfTest['Minimum_Order'] = dfTest['Minimum_Order'].apply(lambda x:int(x.strip('₹')))
dfTest['Average_Cost'] = dfTest['Average_Cost'].apply(lambda x:(x.strip('₹')))
dfTest['Average_Cost'] = dfTest['Average_Cost'].apply(lambda x:int(x.replace(',','')) if x != 'for' else -99)
median = int(dfTest['Average_Cost'].median())
dfTest["Average_Cost"] = np.where(dfTest["Average_Cost"] == 0, median,dfTest['Average_Cost'])
```
